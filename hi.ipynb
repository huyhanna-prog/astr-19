```python
import numpy as np

def get_sin(x):
    return np.sin(x)
```

```python
def get_cos(x):
    return np.cos(x)
```

```python
# Create 1000 entries from 0 to 2
x_values = np.linspace(0, 2, 1000)
sin_values = get_sin(x_values)
cos_values = get_cos(x_values)
```

```python
print(f"{'x':<10} | {'sin(x)':<10} | {'cos(x)':<10}")
print("-" * 35)

# Use a for loop to print the first 10 values in columns
for i in range(10):
    print(f"{x_values[i]:<10.6f} | {sin_values[i]:<10.6f} | {cos_values[i]:<10.6f}")
```
